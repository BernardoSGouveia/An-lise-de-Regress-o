```python
Y = dados.T[0]
X = np.insert(dados.T[1:].T,0,np.array([1 for yi in Y]),1)
```

```python
# Resumo:
nomes = ["Preço da Casa","Impostos","N de banheiros","Tamanho do Lote","Área Construída","Vagas na garagem",
        "N de Cômodos", "N de Quartos", "Idade da Casa", "N de Lareiras"]
def corr(x,y):
    Sxx = sum(((x[i]-np.mean(x))**2 for i in range((len(y)))))
    Syy = sum(((y[i]-np.mean(y))**2 for i in range((len(y)))))
    Sxy = sum(((x[i]-np.mean(x))*(y[i]-np.mean(y)) for i in range((len(y)))))
    return Sxy/((Sxx*Syy)**0.5)
i = 0
M_r = np.zeros((10,10))
for v in dados.T:
    print(str(i+1) +": "+ nomes[i])
    print(f"Média:{np.mean(v):.4f}; Desvio Padrão:{np.std(v):.4f}")
    print(f"Min:{min(v)}; Máx:{max(v)}; Mediana:{np.median(v):.2f}\n")
    j = 0
    for w in dados.T:
        M_r[i][j]= corr(v,w)
        j+=1
    i+=1
for line in M_r:
    linha = ''
    for x in line:
        linha+= f"{abs(x):.3f} "
    print(linha)
```

```python
#Plots (ignorar o primeiro)
fig, axes = plt.subplots(2, 5, figsize=(17, 7))
axes = axes.flatten()

def adj_simples(x,y):
    Sxx = sum(((x[i]-np.mean(x))**2 for i in range((len(y)))))
    Sxy = sum(((x[i]-np.mean(x))*(y[i]-np.mean(y)) for i in range((len(y)))))
    b1 = Sxy/Sxx
    b0 = np.mean(y) - b1*np.mean(x)
    return (b0, b1)
betas = []
for i,x in enumerate(X.T.tolist()):
    axes[i].scatter(x,Y,color="b")
    xi = np.linspace(min(x),max(x))
    b0, b1 = adj_simples(x,Y)
    beta_adj = b0 + b1*xi
    betas.append(b1)
    axes[i].plot(xi,beta_adj, color="r",label=fr"$\beta_1$ = {b1:.3f}")
    axes[i].plot()
    axes[i].set_title(f"{nomes[i]}; r={M_r[0][i]:.3f}")
    axes[i].legend()
plt.show()
```

```python
#Ajuste multivariado
beta_hat = np.linalg.inv(X.T@X)@X.T@Y
Y_hat = X @ beta_hat
e = Y - Y_hat

#Tabela ANOVA:
SQTOT = np.sum((Y - np.mean(Y))**2)
SQRES = np.sum(e**2)
SQREG = SQTOT - SQRES
QMREG = SQREG / len(X.T-1)
QMRES = SQRES / (len(Y) - len(X.T))
F_calc = QMREG / QMRES
print(f"F_calc = {F_calc:.3f}")
R2 = SQREG / SQTOT
R2_adj = 1 - (QMRES / (SQTOT / (len(Y) - 1)))
print(rf"R^2 = {R2:.3f}; R^2_adj = {R2_adj:.3f}")
```

```python
#Teste t
Var_beta = QMRES * np.linalg.inv(X.T@X)
t_calc = beta_hat / np.sqrt(np.diag(Var_beta))
lt = [abs(round(x, 3)) for x in t_calc]
t_table = 2.145 #procurar na tabela t-student para diferentes tamanhos de Y(alpha padrão=0.05)
print(lt)
print("t_tabela= 2.145")
```

```python
def optm(Y,X,ind):
    #Ajuste atual
    beta_hat = np.linalg.inv(X.T@X)@X.T@Y
    Y_hat = X @ beta_hat
    e = Y - Y_hat
    QMRES = np.sum(e**2)/ (len(Y) - len(X.T))
    SQTOT = np.sum((Y - np.mean(Y))**2)
    R2_old =  1 - (QMRES / (SQTOT / (len(Y) - 1)))
    #Ajuste sem a variável xi
    temp_list = X.T.tolist()
    temp_list.pop(ind)
    X = np.array(temp_list).T
    # print(X)
    beta_hat = np.linalg.inv(X.T@X)@X.T@Y
    Y_hat = X @ beta_hat
    e = Y - Y_hat
    QMRES = np.sum(e**2)/ (len(Y) - len(X.T))
    SQTOT = np.sum((Y - np.mean(Y))**2)
    R2_new =  1 - (QMRES / (SQTOT / (len(Y) - 1)))
    return R2_old, R2_new
lt1 = [[i,x] for i,x in enumerate(lt)]
lt1.sort(key= lambda x: x[1])
print(lt1)
for i,x in lt1:
    if x<t_table:
        R2o,R2n = optm(Y,X,i)
        print(f"{nomes[i]}:\n R2_antigo = {R2o:.3f}; R2_adj = {R2n:.3f}")
```

```python
#Otimizando de verdade o ajuste
def remove_column(X,i):
    temp_list = X.T.tolist()
    temp_list.pop(i)
    X1 = np.array(temp_list).T
    return X1

X1 = remove_column(X,lt1[0][0])
for i,t in lt1[1:]:
    E = X.T[i]
    for j,line in enumerate(X1.T):
        if all(line==E):
            index = j
    if t<t_table:
        print(index,nomes[i],X1)
        R2o,R2n = optm(Y,X1,index)
        print(R2o,R2n)
        if R2n >= R2o:
            X1 = remove_column(X1,index)
```
